```python
import requests
import pandas as pd
import time


def fetch_details(game_id,player_id, team_id,context_measure="DEF_FGA"):
    base_url = "https://stats.nba.com/stats/videodetailsasset"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
        "Referer": "https://www.nba.com",
        "Accept": "application/json",
        "Accept-Language": "en-US,en;q=0.9",
        "Origin": "https://www.nba.com"
    }
    
    # Construct params based on your example URL, fill in required ones:
    params = {
        "GameID": game_id,
        "GameEventID":"",
        "PlayerID": player_id,
        "TeamID": team_id,
        "Season": "",
        "SeasonType": "",
        # Other parameters are left blank or defaulted from your example URL
        "AheadBehind": "",
        "CFID": "",
        "CFPARAMS": "",
        "ClutchTime": "",
        "Conference": "",
        "ContextFilter": "",
        "ContextMeasure": context_measure,
        "DateFrom": "",
        "DateTo": "",
        "Division": "",
        "EndPeriod": 0,
        "EndRange": 40800,
        "GROUP_ID": "",
        "GameSegment": "",
        "GroupID": "",
        "GroupMode": "",
        "GroupQuantity": 5,
        "LastNGames": 0,
        "Location": "",
        "Month": 0,
        "OnOff": "",
        "OppPlayerID": "",
        "OpponentTeamID": 0,
        "Outcome": "",
        "PORound": 0,
        "Period": 0,
        "PlayerID1": "",
        "PlayerID2": "",
        "PlayerID3": "",
        "PlayerID4": "",
        "PlayerID5": "",
        "PlayerPosition": "",
        "PointDiff": "",
        "Position": "",
        "RangeType": 0,
        "RookieYear": "",
        "SeasonSegment": "",
        "ShotClockRange": "",
        "StartPeriod": 0,
        "StartRange": 0,
        "StarterBench": "",
        "VsConference": "",
        "VsDivision": "",
        "VsPlayerID1": "",
        "VsPlayerID2": "",
        "VsPlayerID3": "",
        "VsPlayerID4": "",
        "VsPlayerID5": "",
        "VsTeamID": ""
    }

    response = requests.get(base_url, headers=headers, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Request failed with status code {response.status_code}")
        return None


# Example usage with your parameters:
game_id = "0042400402"  # as string to match your example URL

player_id = "1630174"
team_id = "1610612754"
context_measure="DEF_FGA",
video_json = fetch_details(game_id,  player_id, team_id,context_measure=context_measure)
df=pd.DataFrame(video_json['resultSets']['playlist'])

df=df[['gi','ei','dsc']]

df['def_id']=player_id
df['team_id']=team_id


master_record=pd.read_csv('master_record.csv')
master_record.columns
print(df)
```
